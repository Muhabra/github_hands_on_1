```python
def compute_sorori_shinzaemon(n_days=100): #n_days is no of days / default = 0
    num = 1 # no. of 1st grain
    total = 1 # total no. of grains in 1st day
    list_n_grains = [num] # list of grains received each day
    list_total_grains = [total]
    for n in range(n_days - 1):
        num = num*2
        list_n_grains.append(num) #list of grains received each day
        total+=num
        list_total_grains.append(total)#list of total grains received by day
    print("Number of grains in day",n_days,':',num)
    print("Total number of grains by",n_days,"days:",total)
    
    import matplotlib.pyplot as plt #function to draw line graph
    %matplotlib inline
    fig, ax = plt.subplots()
    no_of_days=[x for x in range(1, n_days+1)]
    ax.plot(no_of_days, list_n_grains,label = 'No. of grains per day', color='green',linewidth = '2')
    ax.plot(no_of_days, list_total_grains,label = 'Total no. of grains received by day')
    plt.title("Total Rice received")
    plt.xlabel("Number of days")
    plt.ylabel("Total Number of grains")
    plt.legend()
    plt.show()


compute_sorori_shinzaemon(81)# compute for 81 days
```
