```python
folds = []
folds = [0.00008,]

for i in range(43):
    THICKNESS  =  THICKNESS * 2
    folds.append(THICKNESS)
    folds
```

```python
print(len(folds))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folds)
plt.show()
#### As depicted below, the tickness of the paper increases exponentially
```

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(folds, color='red')
plt.show()
```

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(folds)
plt.show()
```

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(folds, linewidth=3.0, linestyle='--')
plt.show()
```

```python
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.tick_params(labelsize=20)
plt.plot(folds, linewidth=1.0, linestyle='--', color='gold')
plt.show()
```
