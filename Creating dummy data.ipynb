```python
import numpy as np

mean = (-3,0) #The "mean" here is the normal distribution mean
cova = [[1.0, 0.8], #The "Cova" here is abbrivate for covariance
        [0.8, 1.0]]
np.random.seed(0)
random_ndarray = np.random.multivariate_normal(mean,cova,size=500)
print(random_ndarray)
print(random_ndarray.shape)
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
%matplotlib inline

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter Plot")
plt.scatter(random_ndarray[:, 0], random_ndarray[:, 1], c='g', alpha=0.5)
plt.show()
```

```python
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("Histogram of x1")
plt.hist(random_ndarray[:, 0], color = 'b', bins=50)
plt.xlim(right = 3)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("Histogram of x1")
plt.hist(random_ndarray[:, 1], color = 'y', bins=50)
plt.xlim(left = -3)
plt.show()
```

```python
mean1 = (0, -3)
cova = [[1.0, 0.8],
       [0.8, 1.0]]
np.random.seed(0)
first_random_ndarray = np.random.multivariate_normal(mean1,cova,size=500)

plt.xlabel("x1")
plt.ylabel("y2")
plt.title("Scatter Plot")

plt.scatter(random_ndarray[:, 0], random_ndarray[:, 1], label="= 0")
plt.scatter(first_random_ndarray[:, 0], first_random_ndarray[:, 1], label="= 1")
plt.legend()
plt.show()
```

```python
np.random.seed(0)

combined_ndarray = np.concatenate((random_ndarray, first_random_ndarray))
print(combined_ndarray.shape)
print(combined_ndarray)
```

```python
first_ndarray = np.zeros((500, 1))
second_ndarray = np.ones((500, 1))

label = np.concatenate([first_ndarray, second_ndarray], axis=0)

labeled_ndarray = np.concatenate([combined_ndarray, label], axis = 1)
print(labeled_ndarray.shape)
print(labeled_ndarray)
```
