```python
total = house_data.isnull().sum().sort_values(ascending=False)
percent = (house_data.isnull().sum()/house_data.isnull().count()*100).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['total', 'missing_ratio'])
missing_data.head(10)
```

```python
#Deleting variable with 5 or more missing values
df = len(house_data.isnull())
df
```

```python
print("Numbers of columns not containg null values is")
print(len(house_data.columns[house_data.notna().all()]))
```

```python
print("Numbers of columns not containg null values is")
print(len(house_data.columns[house_data.isna().any()]))
```

```python
threshold = 5
missing_features = missing_data.query("total >= {}".format(threshold)).index
missing_features
```

```python
drop_features = house_data.drop(missing_features, axis=1)
display(drop_features)
```

```python
row_drop = drop_features.dropna(axis=0)
row_drop = row_drop.reset_index(drop=True)
row_drop
```

# Terminology Research

Skewness  is a measure of the degree of lopsidedness in the frequency distribution. It is an indicator of lack of symmetry

Kurtosis is a measure of degree of tailedness in the frequency distribution. It is used to denote the horizontal pull on the data


# Confirming distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
for col in house_data[['SalePrice']]:
    print('{:15}'.format(col),
         'Skewness: {:05.2f}'.format(house_data[col].skew()) , 
         '    ' ,
         'Kurtosis: {:06.2f}'.format(house_data[col].kurt())
         )
```

```python
sns.set_theme(); np.random.seed(0)
sns.displot(house_data['SalePrice'], kde=False, bins=10)
```

```python
sns.set_theme(); np.random.seed(0)
target_var = house_data['SalePrice']
sns.distplot(target_var)
plt.show()
```

```python
target_log = target_var.apply(np.log)
sns.displot(target_log, kde=False, bins=10)
```

```python
sns.set_theme(); np.random.seed(0)
log_data = house_data['SalePrice']
sns.distplot(log_data, bins=10)
plt.show()
```

# Correlation coefficient

```python
data_corr = house_data.corr()
plt.figure(figsize=(16, 13))
plt.title('correlation matrix')
sns.heatmap(data_corr, vmax=1, annot=False, cmap='BrBG')
```

```python
# 10 variables that are highly correlate with target variable
corr_target = data_corr[['SalePrice']].sort_values(by=['SalePrice'], ascending=False)
corr_target[1:11]
```
